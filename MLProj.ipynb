```python
from pycaret.datasets import get_data
diabetes = get_data('diabetes')
from pycaret.classification import *
```

```python
# Rename the class variable column to "Outcome"
diabetes.rename(columns={'Class variable': 'Outcome'}, inplace=True)
print(diabetes.columns)
```

```python
s = setup(data=diabetes, target='Outcome')
```

```python
## Function to display model results with labels
def display_model_results(models, labels):
    for model, label in zip(models, labels):
        print(f"Model: {label}")
        print(model)
        print("=" * 50)
```

```python
multiple_models = [create_model(i) for i in ['lr','knn','nb','dt','mlp' , 'svm' , 'gpc','ridge','ada' , 'rf']]
# Set labels for the models
# Define model labels
model_labels = ['Linear Regression', 'K-Nearest Neighbors', 'Naive Bayes', 'Decision Tree', 'Multi-Layer Perceptron', 'Support Vector Machine', 'Gaussian Process', 'Ridge Regression', 'AdaBoost', 'Random Forest']

# Call the function to display model results with labels
display_model_results(multiple_models, model_labels)
```

```python
best_model = compare_models()
```

```python
diabetes_for_pred = diabetes.copy()
```

```python
predictions = predict_model(best_model,diabetes_for_pred)
```
